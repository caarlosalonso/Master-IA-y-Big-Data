```python
# Leer el CSV y convertirlo en un DataFrame
df = pd.read_csv("Product_v6.csv") # Con este método cargamos el dataset y además lo convertimos en un DataFrame
print("Todas las columnas:", df.columns)  # Mostrar todas las columnas

# Seleccionar una columna (ejemplo: 'value') y convertirla en un array de NumPy
array_np = np.array(df["value"])

# Aplicar una operación con NumPy (ejemplo: raíz cuadrada)
resultado_np = np.sqrt(array_np)

# Añadir los resultados como una nueva columna en el DataFrame
df["raiz_cuadrada"] = resultado_np

# Guardar el DataFrame actualizado en un nuevo CSV (opcional)
df.to_csv("datos_modificados.csv", index=False) # evita que Pandas guarde el índice del DataFrame como una columna adicional en el archivo CSV
```

```python
# Crear DataFrames con columnas diferentes
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Luis', 'Pedro', 'Marta']
})

df2 = pd.DataFrame({
    'Salario': [25000, 45000, 35000, 40000],
    'Edad': [28, 34, 22, 40]
})

# Concatenar los DataFrames por columnas (axis=1), aunque no coincidan las columnas
df_concat_diff_cols = pd.concat([df1, df2], axis=1)
print(df_concat_diff_cols)

# Crear dos DataFrames con columnas diferentes
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Luis', 'Pedro', 'Marta']
})

df2 = pd.DataFrame({
    'Salario': [25000, 45000, 35000, 40000]
})

# Concatenar los DataFrames por columnas (axis=1)
df_concat_col = pd.concat([df1, df2], axis=1)
print(df_concat_col)
```
