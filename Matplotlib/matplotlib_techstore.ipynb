## Setup

Importación de librerías necesarias:


````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

## Carga de Datos

Ejecuta esta celda para cargar el dataset:


````python
# Generar dataset de ventas TechStore 2024
np.random.seed(42)
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

ventas = pd.DataFrame({
    'mes': meses,
    'laptops': np.random.randint(80, 200, 12),
    'smartphones': np.random.randint(150, 400, 12),
    'tablets': np.random.randint(50, 150, 12),
    'accesorios': np.random.randint(200, 500, 12)
})

print("Dataset TechStore cargado correctamente")
print(ventas.head())
print(f"\nForma: {ventas.shape}")
````

## Ejercicio 1: Gráfico de línea - Evolución mensual

**Objetivo:** Crear un gráfico de línea con múltiples series

Queremos ver la evolución de ventas de los 4 productos a lo largo del año. Cada producto debe tener su propia línea con color, marcador y etiqueta.



### Tareas

- Crear figura y axes con plt.subplots()
- Dibujar una línea por producto con ax.plot()
- Cada línea con color diferente y marcador
- Añadir título, etiquetas de ejes y leyenda
- Añadir grid para facilitar lectura


````python
# EJERCICIO 2: Gráfico de Línea - Evolución de Ventas
# Dataset: ventas (DataFrame)

fig, ax = plt.subplots(figsize=(12, 6))

# Dibujar línea para cada producto
ax.plot(ventas['mes'], ventas['laptops'], marker='o', lw=2,
        color='#2563EB', label='Laptops')

# TODO: Añadir smartphones (color rojo #DC2626, marker='s')
ax.plot(ventas['mes'], ventas['smartphones'], marker='s', lw=2,
        color='#DC2626', label='Smartphones')

# TODO: Añadir tablets (color verde #059669, marker='^')
ax.plot(ventas['mes'], ventas['tablets'], marker='^', lw=2,
        color='#059669', label='Tablets')

# TODO: Añadir accesorios (color naranja #D97706, marker='D')
ax.plot(ventas['mes'], ventas['accesorios'], marker='D', lw=2,
        color='#D97706', label='Accesorios')

# Personalización
ax.set_title('Ventas Mensuales por Producto - TechStore 2024',
             fontsize=16, fontweight='bold')
ax.set_xlabel('Mes', fontsize=12)
ax.set_ylabel('Unidades vendidas', fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## Ejercicio 2: Scatter plot - Relación entre variables

**Objetivo:** Visualizar correlación con color como tercera dimensión

Queremos ver si hay relación entre las ventas de laptops y smartphones. Además, usaremos el color para representar el mes (0=Enero, 11=Diciembre).



### Tareas

- Crear scatter plot con laptops en X y smartphones en Y
- Colorear puntos según el mes (usar np.arange(12))
- Añadir barra de color con plt.colorbar()
- Personalizar con título y etiquetas


````python
# EJERCICIO 3: Scatter Plot - Relación Laptops vs Smartphones

fig, ax = plt.subplots(figsize=(10, 6))

# TODO: Crear scatter plot
# - X: ventas['laptops']
# - Y: ventas['smartphones']
# - Color: np.arange(12) para representar mes
# - Usar cmap='viridis', s=100, edgecolors='black'
scatter = ax.scatter(ventas['laptops'], ventas['smartphones'],
                     c=np.arange(12), cmap='viridis', s=100, edgecolors='black')

# TODO: Añadir colorbar
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Mes (0=Ene, 11=Dic)', rotation=270, labelpad=15)

# Personalización
ax.set_title('Relación Laptops vs Smartphones',
             fontsize=16, fontweight='bold')
ax.set_xlabel('Ventas Laptops', fontsize=12)
ax.set_ylabel('Ventas Smartphones', fontsize=12)

plt.tight_layout()
plt.show()
````

## Ejercicio 3: Subplots 2x2 - Dashboard

**Objetivo:** Crear una cuadrícula de gráficos para comparar productos

Crea un dashboard con 4 gráficos de barras, uno por producto, mostrando las ventas mensuales de cada uno.



### Tareas

- Crear figura con subplots(2, 2)
- Un gráfico de barras por producto
- Acceder a cada axes con axes[fila, columna]
- Añadir título general con fig.suptitle()


````python
# EJERCICIO 4: Subplots 2x2 - Dashboard por Producto

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

productos = ['laptops', 'smartphones', 'tablets', 'accesorios']
colores = ['#2563EB', '#DC2626', '#059669', '#D97706']

# TODO: Crear un gráfico de barras para cada producto
# Usar axes[fila, col].bar(ventas['mes'], ventas[producto], color=color)
# Añadir título a cada subplot
for i, producto in enumerate(productos):
    row, col = divmod(i, 2)
    axes[row, col].bar(ventas['mes'], ventas[producto], color=colores[i])
    axes[row, col].set_title(f'Ventas de {producto.capitalize()}',
                             fontsize=14, fontweight='bold')
    axes[row, col].set_xlabel('Mes', fontsize=10)
    axes[row, col].set_ylabel('Unidades vendidas', fontsize=10)
    axes[row, col].grid(True, alpha=0.3)

# Título general
fig.suptitle('Ventas por Producto - TechStore 2024',
             fontsize=18, fontweight='bold')

plt.tight_layout()
plt.show()
````

## Ejercicio 4: Personalización avanzada

**Objetivo:** Añadir líneas de referencia y anotaciones

Mejora el gráfico de smartphones añadiendo una línea horizontal con la media y una anotación señalando el mes con más ventas.



### Tareas

- Crear gráfico de barras de smartphones
- Añadir línea horizontal con la media (axhline)
- Encontrar el mes con máximas ventas
- Añadir anotación con flecha señalando el máximo


````python
# EJERCICIO 5: Personalización con Media y Anotación

fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de barras
ax.bar(ventas['mes'], ventas['smartphones'],
       color='#DC2626', alpha=0.8,
       edgecolor='black', linewidth=0.5)

# TODO: Calcular y dibujar línea de media
media_smartphones = ventas['smartphones'].mean()
ax.axhline(media_smartphones, color='blue', linestyle='--',
           label=f'Media: {media_smartphones:.1f}')

# TODO: Encontrar mes con máximas ventas
max_mes = ventas.loc[ventas['smartphones'].idxmax(), 'mes']
max_valor = ventas['smartphones'].max()
ax.annotate(f'Máximo: {max_valor} en {max_mes}',
            xy=(max_mes, max_valor), xytext=(max_mes, max_valor + 20),
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=10, fontweight='bold', ha='center')

# TODO: Añadir anotación con flecha
min_mes = ventas.loc[ventas['smartphones'].idxmin(), 'mes']
min_valor = ventas['smartphones'].min()
ax.annotate(f'Mínimo: {min_valor} en {min_mes}',
            xy=(min_mes, min_valor), xytext=(min_mes, min_valor - 25),
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=10, fontweight='bold', ha='center')


# Personalización
ax.set_title('Ventas de Smartphones 2024', fontsize=16, fontweight='bold')
ax.set_xlabel('Mes', fontsize=12)
ax.set_ylabel('Unidades', fontsize=12)
ax.legend(fontsize=11)

plt.tight_layout()
plt.show()
````
